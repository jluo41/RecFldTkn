# Split

```python
##################################
CaseSplitConfig = 'split_io_tvt_disease_cohort'
##################################
import numpy as np  


import inspect


def conduct_spliting_ds_case(df_case, case_id_columns):
    
    conditions = [
        df_case['Test'],
        df_case['Valid'],
        df_case['Train']
    ]
    choices = ['Test', 'Valid', 'Train']
    df_case['TVT'] = np.select(conditions, choices, default=np.nan)

    conditions = [
        df_case['In'],
        df_case['Out'],
        # df_case['Train']
    ]
    choices = ['In', 'Out']
    df_case['InOut'] = np.select(conditions, choices, default=np.nan)
    df_case['disease'] = 'T' + df_case['DiseaseType'].astype(str) + 'D'

    split_name_columns = ['InOut', 'TVT', 'disease', 'cohort']
    df_case['splitgroup'] = df_case[split_name_columns].agg('_'.join, axis=1)

    splitname_to_dfcase = {}
    for splitname, df in df_case.groupby('splitgroup'):
        # print(splitname, df.shape)
        splitname_to_dfcase[splitname] = datasets.Dataset.from_pandas(df[case_id_columns].reset_index(drop=True))
    return splitname_to_dfcase    

conduct_spliting_ds_case.fn_string = inspect.getsource(conduct_spliting_ds_case)

from recfldtkn.loadtools import convert_variables_to_pystirng

prefix = [
    'import datasets',
    'import pandas as pd', 
    'import numpy as np']
fn_variables = [conduct_spliting_ds_case]
pycode = convert_variables_to_pystirng(fn_variables = fn_variables, prefix = prefix)
pypath = os.path.join(SPACE['CODE_FN'], 'fn_learning', f'{CaseSplitConfig}.py')
print(pypath)
if not os.path.exists(os.path.dirname(pypath)): os.makedirs(os.path.dirname(pypath))
with open(pypath, 'w') as file: file.write(pycode)
```

```python
ds_case_dict = conduct_spliting_ds_case(df_case, case_id_columns)
# for k, v in ds_case_dict.items():
#     print(k, v.shape)

ds_case_dict
```
