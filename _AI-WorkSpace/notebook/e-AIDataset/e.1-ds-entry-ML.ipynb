# EntryFn-MLUniPred


Binary Prediction, Numeric Prediction. Only one dimension to predict.


Multi-Pred: predict dim1, dim2, dim3, dim4, xxx.

```python
TASK = 'MLUniPred'
ENTRY_METHOD = f'EntryFor{TASK}'

EntryArgs = {
    'INPUT_CF': 'InvLevel.DemoInvRx.ObsPnt',
    'OUTPUT_CF': 'InvLevel.Egm.Af1W',
    "OUTPUT_TKN": 'Af1wEdu:InvEgm',
}
```

```python
import inspect
from recfldtkn.loadtools import convert_variables_to_pystirng

def fn_entry_method_for_finaldata(dataset, cf_to_CFVocab, EntryArgs):
    from scipy.sparse import csr_matrix
    import numpy as np
    import itertools
    
    def filter_tkn(key, value, tid2tkn_filter):
        index = np.array([bool(tid in tid2tkn_filter) for tid in key])
        selected = list(np.array(value)[index])
        return selected 
    

    # Input feaures. 

    INPUT_CF = EntryArgs['INPUT_CF'] # 
    CF_vocab = cf_to_CFVocab[INPUT_CF]
    tid2tkn = CF_vocab['input_ids']['tid2tkn']
    num_features = len(tid2tkn)

    tid2tkn_filter = EntryArgs.get('tid2tkn_filter', None)
    input_ids_name = f'{INPUT_CF}.input_ids'
    input_wgts_name = f'{INPUT_CF}.input_wgts'

    if tid2tkn_filter:
        col_indices = list(itertools.chain(*[          filter_tkn(tid, tid, tid2tkn_filter)   for i,   tid in enumerate(dataset[input_ids_name])]))
        row_indices = list(itertools.chain(*[[i] * len(filter_tkn(tid, tid, tid2tkn_filter))  for i,   tid in enumerate(dataset[input_ids_name])]))
        data        = list(itertools.chain(*[          filter_tkn(tid, wgt, tid2tkn_filter)   for tid, wgt in zip(dataset[input_ids_name], dataset[input_wgts_name])]))
    else:
        col_indices = list(itertools.chain(*[          tid  for i,   tid in enumerate(dataset[input_ids_name])]))
        row_indices = list(itertools.chain(*[[i] * len(tid) for i,   tid in enumerate(dataset[input_ids_name])]))
        data        = list(itertools.chain(*[          wgt  for tid, wgt in zip(dataset[input_ids_name], dataset[input_wgts_name])]))
       
    sparse_matrix_value = (data, (row_indices, col_indices))
    shape = (len(dataset), num_features)
    X = csr_matrix(sparse_matrix_value, shape=shape)


    # Output labels.
    OUTPUT_CF = EntryArgs['OUTPUT_CF']
    if OUTPUT_CF in cf_to_CFVocab:
        OUTPUT_TKN = EntryArgs['OUTPUT_TKN']
        output_tknidx = cf_to_CFVocab[OUTPUT_CF]['input_ids']['tkn2tid'][OUTPUT_TKN]
        data = [dict(zip(tid, wgt)).get(output_tknidx, 0)  for tid, wgt in zip(dataset[f'{OUTPUT_CF}.input_ids'], dataset[f'{OUTPUT_CF}.input_wgts'])]
        labels = np.array(data).astype(int)
        dataset = {'X': X, 'Y': labels}
    else:
        dataset = {'X': X}
    return dataset

fn_entry_method_for_finaldata.fn_string = inspect.getsource(fn_entry_method_for_finaldata)
prefix = ['import pandas as pd', 'import numpy as np']
fn_variables = [fn_entry_method_for_finaldata]
pycode = convert_variables_to_pystirng(fn_variables = fn_variables, prefix = prefix)
pypath = os.path.join(SPACE['CODE_FN'], 'fn_learning', f'{ENTRY_METHOD}.py')
print(pypath)
if not os.path.exists(os.path.dirname(pypath)): os.makedirs(os.path.dirname(pypath))
with open(pypath, 'w') as file: file.write(pycode)
```
